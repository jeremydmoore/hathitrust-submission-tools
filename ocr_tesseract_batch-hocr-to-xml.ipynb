```python
dir_path_list = sorted([x for x in root_dir.iterdir() if x.is_dir()])
len(dir_path_list)
```

```python
for dir_path in dir_path_list:
    print(f'Processing {dir_path.name} . . .')
    tif_path_list = sorted(dir_path.glob('*.tif'))
    number_of_tif = len(tif_path_list)
    for i, tif_path in enumerate(tif_path_list):
        print(f'\t\tOCRing {i+1} of {number_of_tif} . . . . . . . . {number_of_tif-i} to go')
        !tesseract {str(tif_path)} {tif_path.parent.joinpath(tif_path.stem)} 1>/dev/null 2>&1
        !tesseract {str(tif_path)} {tif_path.parent.joinpath(tif_path.stem)} hocr 1>/dev/null 2>&1
        hocr_file_name = tif_path.stem + '.hocr'
        hocr_file_path = tif_path.parent.joinpath(hocr_file_name)
        new_hocr_file_name = tif_path.stem + '.xml'
        new_hocr_file_path = tif_path.parent.joinpath(new_hocr_file_name)
        hocr_file_path.replace(new_hocr_file_path)
```

```python
for tif in *.tif; do stub=$(basename $tif .tif); echo OCRing $tif; tesseract $tif $stub 1>/dev/null 2>&1; tesseract $tif $stub hocr 1>/dev/null 2>&1; mv $stub.hocr $stub.xml; done
```
